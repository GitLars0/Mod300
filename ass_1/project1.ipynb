```python
import matplotlib.pyplot as plt
import numpy as np

class FancyWavePacket:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __call__(self, x):
        return np.sin(self.b * x) * np.exp(-self.a * x * x)

    def forward_difference(self, x, h):
        return (self(x + h) - self(x)) / h

    def central_difference(self, x, h):
        return (self(x + h) - self(x - h)) / (2 * h)

    #As we can see thhe forward and central difference get very similar values

    def exact_derivative(self, x):
        return self.b * np.cos(self.b * x) * np.exp(-self.a * x * x) - 2 * self.a * x * np.sin(self.b * x) * np.exp(-self.a * x * x)

    def plot_error(self, x):
        step_sizes = np.logspace(-16, 0, 100)
        exact_value = self.exact_derivative(x)
        forward_errors = [abs(self.forward_difference(x, h) - exact_value) for h in step_sizes]
        central_errors = [abs(self.central_difference(x, h) - exact_value) for h in step_sizes]


        plt.scatter(step_sizes, forward_errors, color='red', label='Forward Difference Error')
        plt.scatter(step_sizes, central_errors, color='blue', label='Central Difference Error')
        plt.xscale('log')
        plt.yscale('log')
        plt.xlabel('Step Size (h)')
        plt.ylabel('Absolute Error')
        plt.title(f'Error of Finite Difference Approximations at x = {x}')
        plt.legend()
        plt.grid(True, which="both", linestyle="--")
        plt.show()

WP1 = FancyWavePacket(0.1, 2)
WP2 = FancyWavePacket(0.1, 10)
x = 1
h = 0.001

print(f"WP1(x): {WP1(x)}")
print(f"WP2(x): {WP2(x)}")
print(f"WP1 forward difference derivative at x: {WP1.forward_difference(x, h)}")
print(f"WP2 forward difference derivative at x: {WP2.forward_difference(x, h)}")
print(f"WP1 central difference derivative at x: {WP1.central_difference(x, h)}")
print(f"WP2 central difference derivative at x: {WP2.central_difference(x, h)}")

WP1.plot_error(x)
```

```python
import numpy as np
class duple:
    """
    Class for automatic differentiation
    top: function value
    bottom: derivative of function
    """
    def __init__(self,top, bottom=0):
        self.top = top
        self.bottom = bottom
    def __add__(self, d): # u+v, uâ+vâ
        return duple(self.top + d.top, self.bottom + d.bottom ) #Function for adding
    def __repr__(self):
        return "["+str(self.top)+","+str(self.bottom)+"]" 
    def __str__(self):
        return "["+str(self.top)+","+str(self.bottom)+"]" 
    def __neg__(self):
        return "["+str(-self.top)+ ","+str(-self.bottom)+"]" #Function for handling negative x
    def __pos__(self):
        return "["+str(self.top)+","+str(self.bottom)+"]" #Function for handling +x
    def __sub__(self, d):
        return duple(self.top - d.top, self.bottom - d.bottom) #Function for subtracting
    def __mul__(self, d):
        return duple(self.top * d.top, self.bottom*d.top + self.top*d.bottom) #Function for multiplication
    def __truediv__(self, d):
        return duple(self.top/d.top,( self.bottom*d.top - self.top*d.bottom)/d.top**2) #Function for division


x=duple(1,2)
y=duple(3,4)
z=x+y
print(z)
print(-x)
print(+x)
print(x-y)

x=1.2 # just for making it easy to use other values
One=duple(1.,0.) # derivative of a constant is zero
X=duple(x,1) # derivative of x with respect to x is 1
print("auto diff=",X*X*X)
print("analytical=",x*x*x, 3*x*x)
print("auto diff=", One/(One+X))
print("analytical=", 1/(1+x),-1/(1+x)**2)
Exp=duple(np.exp(x),np.exp(x)) # derivative of exp(x) is exp(x)
Sin=duple(np.sin(x),np.cos(x)) # derivative of sin(x) is cos(x)
print("auto diff=", Exp*Sin)
print("analytical=", np.exp(x)*np.sin(x), np.exp(x)*np.sin(x)+np.exp(x)*np.cos(x))

class dfunction:
    def __init__(self, f, df):
        self.f = f
        self.df = df
    def __call__(self,d: duple): # kernel rule sending in and returning duple
        return duple(self.f(d.top), self.df(d.top)*d.bottom)
    

def f(x):
    return np.log(x)
def df(x):
    return 1/x
x=3.1
One=duple(1.,0.) # derivative of a constant is zero
X=duple(x,1) # derivative of x with respect to x is 1
Log=dfunction(f,df) #note no arguments to f - we are passing the function
Exp=dfunction(np.exp,np.exp) # same here
print("auto diff ", Log(One+Exp(X)))
print("analytical ", np.log(1+np.exp(x)),np.exp(x)/(1+np.exp(x)))

a = 0.1
b = 10
x = 1
def wave_function(x): #Function for equation one in project file.
    return np.sin(b*x) * np.exp(-a*x**2)
def wave_func_derivative(x):
    return b * np.cos(b*x)*np.exp(-a*x**2) - 2*a*x*np.sin(b*x)*np.exp(-a*x**2) #Derivative of equation 
    
wave = dfunction(wave_function, wave_func_derivative)
result = wave(duple(x, 1))
analytic = wave_func_derivative(x)

print(f"f(x) = {result.top}")
print(f"f'(x) (numerical) = {result.bottom}")
print(f"f'(x) (analytical) = {analytic}")
```

- Automatic differentiation is quite important in the field of machine learning. It is used for traning netrual networks to quickly calculate derivatives. It is also used in scientific computing to do simulations where models require the calculations of derivatives. 

- Strenghts of automatic differentiation: Accuracy is one of the of strenghts of automatic differentiation, unlike numerical differation which can suffer from round-off errors. Another strenght is efficiency, automatic differentiation is often faster than numerical, specially for complex functions. 

- Weaknesses: Implementing automatic differentiation can be more complex, especially for large systems. Also in some cases automatic differentiation can introduce computational overhead in memory usage.
