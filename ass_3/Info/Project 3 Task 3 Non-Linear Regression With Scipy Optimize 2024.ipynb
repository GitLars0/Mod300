```python
# Classwork, Example with Power Law Rheological model to fit Viscometer data
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

# Define the mapping Powerlaw function
def Power(x, a, b):
	return a * x**b

# Define the measured viscometer data, input (x) and output (y)
# For practice (classwork)
x=np.array([5.10, 10.2, 170.3,340.6,510.9,1021.8])
y=np.array([8.5, 9.6,19.2, 25.6, 32.1,45.9])

# For your project task
# import files and define x and y
#df = pd.read_excel('filename')
#x = df['']
#y= df['']

# Curve fit
popt,_ = curve_fit(Power, x, y)

# Summarize the parameter values/ Unpack the parameters
a, b = popt

#Print the equation of the Power Law model 
print('y = %.5fx^%.5f' % (a, b))


# Powerlaw model prediction (y_model) for the given shear rate (x)
y_model = Power(x, a, b)

# Create a line plot for the mapping function (Power Law model)
plt.plot(x, y_model, '--', color='red')
# Plot input vs output dataset in the scatter plot
plt.scatter(x, y)

plt.xlabel('Shear rate, 1/s')
plt.ylabel('Shear stress, lbf/100ft^2')
plt.title('Power Law Model')
plt.show()

# Model accuracy performance Analysis 
print('R2 =', r2_score(y,y_model))
```
